```python
import pandas as pd

# Carregando o arquivo CSV
file_path = 'analise_simple_cnn1.csv'
data = pd.read_csv(file_path)

# Verificando as primeiras linhas do arquivo para entender a estrutura dos dados
data.head()
```

```python
import pandas as pd

# Carregando o arquivo CSV
file_path = 'analise_simple_cnn2.csv'
data_2 = pd.read_csv(file_path)

# Verificando as primeiras linhas do arquivo para entender a estrutura dos dados
data.head()
```

```python
import matplotlib.pyplot as plt

# Definindo os dados para os grficos
epochs = data.index
loss_training = data['loss_traning']
loss_validation = data['loss_validation']
accuracy_training = data['acuracia_traning']
accuracy_validation = data['acuracia_val']

# Criando os grficos
plt.figure(figsize=(12, 5))

# Grfico de perda (loss)
plt.subplot(1, 2, 1)
plt.plot(epochs, loss_training, label='Treinamento', color='blue')
plt.plot(epochs, loss_validation, label='Validao', color='red')
plt.title('Perda durante o Treinamento e Validao')
plt.xlabel('pocas')
plt.ylabel('Perda')
plt.legend()

# Grfico de acurcia
plt.subplot(1, 2, 2)
plt.plot(epochs, accuracy_training, label='Treinamento', color='blue')
plt.plot(epochs, accuracy_validation, label='Validao', color='red')
plt.title('Acurcia durante o Treinamento e Validao')
plt.xlabel('pocas')
plt.ylabel('Acurcia')
plt.legend()

# Mostrar os grficos
plt.tight_layout()
plt.show()
```
